```python
#Clean the cells - Empty spaces, all lower cases 
type_columns = ["food_type", "other_things_type", "subscriptions_type", "family_type"]

for col in type_columns:
    data[col] = data[col].str.strip().str.lower().str.replace(" ", "_").str.replace(r'[^\w]', '', regex=True)
```

```python
#Lets model the data into 4 columns: Date, Category, Amount, Detail
detailed = {
    "food": "food_type",
    "subscriptions": "subscriptions_type",
    "other_things": "other_things_type",
    "family": "family_type"
}
simple = [
    "transport",
    "housing_utilities"
]

tidy_rows = []

for amount_col, detail_col in detailed.items():
    temp = data[["date", amount_col, detail_col]].copy()
    temp = temp.rename(columns={
        amount_col: "amount",
        detail_col: "detail"
    })
    temp["category"] = amount_col
    tidy_rows.append(temp)


for amount_col in simple:
    temp = data[["date", amount_col]].copy()
    temp = temp.rename(columns={amount_col: "amount"})
    temp["detail"] = None
    temp["category"] = amount_col
    tidy_rows.append(temp)

tidy = pd.concat(tidy_rows, ignore_index=True)

#Delete empty rows
tidy = tidy.dropna(subset=['amount'])
tidy = tidy[tidy['amount'] > 0]
```
